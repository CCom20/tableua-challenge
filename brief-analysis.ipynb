```python
import pandas as pd
```

```python
data = pd.read_csv("JC-202101-citibike-tripdata.csv")
```

```python
data.head()
```

```python
most_common_routes = data.groupby(["start station name", "end station name"]).size().reset_index()
```

```python
most_common_routes
```

```python
most_common_routes = most_common_routes.rename(columns={0: "count"})
```

```python
most_common_routes = most_common_routes[most_common_routes["count"] >= 100]
```

```python
most_common_routes
```

```python
routes_cleaned = pd.merge(most_common_routes, data, on="start station name")
```

```python
routes_cleaned = routes_cleaned.drop_duplicates(subset=["start station name"])
```

```python
routes_cleaned = routes_cleaned[["start station name", "end station name_x", "count", "start station latitude", "start station longitude", "end station latitude", "end station longitude"]]
```

```python
routes_cleaned = routes_cleaned.rename(columns={"end station name_x": "end station name"})
```

```python
routes_cleaned.to_csv("most-common-routes.csv")
```
